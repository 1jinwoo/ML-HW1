```python
import numpy as np
import matplotlib.pyplot as plt
```

# 3 (ii)

```python
def generate_gaussian_hist(d):
    # Generate 10,000 datapoints
    mean = np.zeros(d)
    cov = np.identity(d)
    pts = np.random.multivariate_normal(mean, cov, size=10000)
    
    # Take squared length of each datapoints
    len_sq = [np.linalg.norm(pt) ** 2 for pt in pts]
    
    # Plot the histogram
    plt.hist(len_sq)

# Plotting    
D = [1, 2, 3, 5, 10, 50, 100]
for d in D:
    generate_gaussian_hist(d)
plt.title('Histogram of squared lengths of sampled datapoints\ndrawn from various dimensions of Gaussian')
plt.xlabel('Vector length squared')
plt.ylabel('Occurrences')
plt.legend(D)
plt.show()
```
